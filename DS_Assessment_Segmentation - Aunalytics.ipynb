### Data Visualizations

### Heatmap

```python
plt.figure(figsize = (10, 6))

sns.heatmap(df_train.corr(), annot = True)
```

From the heatmap we can see that the numerical features have no relation with each other, hence we can have keep all features

### Plots for analyzing different features

### Workclass vs Income

```python
plt.figure(figsize = (10,6))
sns.countplot(df_train['workclass'],hue=df_train['class'])
plt.title('Workclass Vs Income Countplot', size = 15)
plt.xticks(rotation = 20)
```

From the plot, an interesting analysis is that highest number of people earning above 50k is from Private workclass also people earning <50k are also higher in private workclass.

comparitevely, its also seen that Federal-gov and self employed not income people year less than 50k

### Education Vs Income

```python
plt.figure(figsize = (15,6))
sns.countplot(df_train['education'],hue=df_train['class'])
plt.title('Education Vs Income Countplot', size = 15)
plt.xticks(rotation = 20)
```

Another interesting analysis is that we can see higher the education, higher the earning. people with higher education earn more than 50k. 

### Marital Status Vs Income

```python
plt.figure(figsize = (15,6))
sns.countplot(df_train['marital-status'],hue=df_train['class'])
plt.title('Marital_status Vs Income Countplot', size = 15)
plt.xticks(rotation = 20)
```

The above graph shows, married people earn more than 50k than people in other marital status.

### Age Vs Income

```python
plt.figure(figsize = (10,5))
sns.catplot(x="class",y="age",kind='violin',data=df_train)
plt.title("Distribution of age for different Income class", size = 13)
```

From the violin plot we can understand the probability and distribution of the data. The age distribution shows that people between age 35-43 earn more than 50k on higher probability, than youngesters. 

### Gender Vs Income

```python
plt.figure(figsize = (8,6))
sns.barplot(x="class",y="education-num",hue = 'sex',data=df_train)
plt.title('Distribution of Income on Gender and Age', size = 13)
```

Its interesting to see that number of female who earns greater than 50k is higher than males

### Native country Vs Income

```python
native_country = round(pd.crosstab(df_train['native-country'], df_train['class']).div(pd.crosstab(df_train['native-country'], df_train['class']).apply(sum,1),0),2)
pl = native_country.plot(kind ='bar', figsize = (20,12), color = {'red','blue'})
pl.set_xlabel('Native country')
pl.set_ylabel('Population')
pl.set_title('Distribution of Native country vs Income')
```

The native country plot shows that people living towards europe, Asia earn more than other countries

### Pairplot

### Another distribution plot of age

```python
plt.figure(figsize = (12,8))
df_50 = df_train[df_train['class'] == ' <=50K']
df_gt50 = df_train[df_train['class'] == ' >50K']
sns.distplot(df_50['age'], kde_kws={"label": "<=$50K"})
sns.distplot(df_gt50['age'], kde_kws={"label": ">$50K"})
```

This plot proves the already stated point that people between the age 35-50 approx earns more than 50k than youngsters and olders.
